```python
import csv
```

```python
data=open('example.csv',encoding='utf-8')
csv_data=csv.reader(data)
data_lines=list(csv_data)
```

```python
data_lines[:3]
```

```python
for line in data_lines[:5]:
    print(line)
```

```python
len(data_lines)
```

```python
all_emails=[]

for line in data_lines[1:]:
    all_emails.append(line[3])
    
all_emails
```

```python
full_names=[]

for line in data_lines[1:]:
    full_names.append(line[1]+' '+line[2])

full_names
```

```python
file_to_output = open('to_save_file.csv','w',newline='')
csv_writer=csv.writer(file_to_output,delimiter=',')
csv_writer.writerow(['full_name','email'])
zipped_file=zip(full_names[0:500],all_emails[0:500])
csv_writer.writerows(zipped_file)
file_to_output.close()
```

```python
file_to_output = open('to_save_file.csv','a',newline='')
csv_writer=csv.writer(file_to_output,delimiter=',')
csv_writer.writerow(['----------| ',' |---------'])
zipped_file=zip(full_names[500:],all_emails[500:])
csv_writer.writerows(zipped_file)
file_to_output.close()
```
